```python
import numpy as np
from PIL import Image

def rgb_to_bgr(image_path, save_path="bgr_output.png"):
    img = np.array(Image.open(image_path).convert("RGB"))
    bgr = img[:, :, ::-1]  # Reverse channel order: RGB → BGR
    Image.fromarray(bgr).save(save_path)
    print(f"Saved BGR image to: {save_path}  |  Shape: {bgr.shape}")
    return bgr

bgr = rgb_to_bgr("Fig.5.jpg") 
```
